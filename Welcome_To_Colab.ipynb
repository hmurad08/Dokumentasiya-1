```
import pandas as pd

# Verilən ədədi dəyərlər
data = [
    165, 148, 176, 189, 157, 150, 158,
    143, 155, 153, 158, 186, 161, 175,
    166, 145, 154, 171, 169, 199, 149,
    199, 177, 174, 172, 166, 167, 161,
    171, 161, 152, 152, 161, 198, 161,
    171, 181, 188, 178, 171, 168, 193,
    156, 153, 162, 151, 161, 135, 172,
    171, 197, 172, 186, 152, 175
]

# 1. Variyasiya sırası
variation_order = sorted(data)
print("Variyasiya sırası:", variation_order)

# 2. Tezlik və nisbi tezliklərin statistik sırası
frequency = pd.Series(data).value_counts().sort_index()
relative_frequency = frequency / len(data)

# Tezlik və nisbi tezlik cədvəli
frequency_df = pd.DataFrame({
    'Tezlik': frequency,
    'Nisbi Tezlik': relative_frequency
}).reset_index().rename(columns={'index': 'Dəyər'})

print("\nTezlik və nisbi tezliklərin statistik sırası:\n", frequency_df)

# 3. Interval statistik sırası (qruplaşdırma)
# Qrupların intervalını təyin edirik
bins = pd.cut(data, bins=range(130, 201, 10))
interval_frequency = pd.value_counts(bins)

# Interval statistik cədvəli
interval_df = pd.DataFrame({
    'Interval': interval_frequency.index,
    'Tezlik': interval_frequency.values
}).reset_index(drop=True)

print("\nInterval statistik sırası:\n", interval_df)
```

```
import numpy as np
import matplotlib.pyplot as plt

# Verilmiş dəyərlər
X = np.array([5, 10, 15, 20, 25])
n_i = np.array([3, 9, 18, 14, 16])

# Kümülatif sıxlıq funksiyası
F_x = np.cumsum(n_i) / np.sum(n_i)

# Empirik paylanma funksiyasının qrafiki
plt.step(X, F_x, where='post', label='Empirik Paylanma Funksiyası')

# Qrafik tənzimləmələri
plt.xlabel('X')
plt.ylabel('F(x)')
plt.title('Empirik Paylanma Funksiyası')
plt.grid(True)
plt.legend()

# Qrafiki göstər
plt.show()
```

```
import matplotlib.pyplot as plt
import numpy as np

# Verilmiş məlumatlar
x_intervals = ['5-10', '10-15', '15-20', '20-25', '25-30']
ni = [3, 9, 18, 14, 16]
total_frequency = sum(ni)

# 1) Nisbi tezlikləri tapmaq
relative_frequencies = [freq / total_frequency for freq in ni]

# 2) Diskret paylanma funksiyasını (PMF) müəyyən etmək
def pmf(x):
    if x == '5-10':
        return relative_frequencies[0]
    elif x == '10-15':
        return relative_frequencies[1]
    elif x == '15-20':
        return relative_frequencies[2]
    elif x == '20-25':
        return relative_frequencies[3]
    elif x == '25-30':
        return relative_frequencies[4]
    else:
        return 0  # Tanınmayan intervallar üçün

# 3) Moda və medianı tapmaq
max_frequency = max(ni)
mode_interval = x_intervals[ni.index(max_frequency)]

# Median tapmaq üçün kumulyativ tezlikləri hesablamaq
cumulative_frequencies = np.cumsum(ni)
median_index = np.searchsorted(cumulative_frequencies, total_frequency / 2)
median_interval = x_intervals[median_index]

# Nisbi tezliklərin histoqramını qurmaq
plt.figure(figsize=(8, 5))
plt.bar(x_intervals, relative_frequencies, color='skyblue', edgecolor='black')
plt.xlabel('Interval')
plt.ylabel('Nisbi Tezlik')
plt.title('Nisbi Tezliklərin Histoqramı')
plt.show()

# Nəticələri göstərmək
print("Diskret paylanma funksiyası:")
for interval in x_intervals:
    print(f"P(X = '{interval}') = {pmf(interval)}")

print("Moda intervalı:", mode_interval)
print("Median intervalı:", median_interval)
```
